```python
import random

def create_grid(rows, cols, alive_cells):
    grid = [[0 for _ in range(cols)] for _ in range(rows)] # Initialize all grid cells as dead (0)
    for r, c in alive_cells:
        grid[r][c] = 1
    return grid

def count_neighbors(grid, r, c):
    rows, cols = len(grid), len(grid[0])
    count = 0
    for dr in [-1, 0, 1]:
        for dc in [-1, 0, 1]:
            if dr == 0 and dc == 0:
                continue
            nr, nc = r + dr, c + dc
            if 0 <= nr < rows and 0 <= nc < cols:
                count += grid[nr][nc]
    return count

def step(grid):
    rows, cols = len(grid), len(grid[0])
    new_grid = [[0 for _ in range(cols)] for _ in range(rows)]

    for r in range(rows):
        for c in range(cols):
            neighbors = count_neighbors(grid, r, c)
            
            if grid[r][c] == 1:  # alive cell rules
                match neighbors:
                    case 0 | 1:
                        new_grid[r][c] = 0  
                    case 2 | 3:
                        new_grid[r][c] = grid[r][c]  
                    case _:
                        new_grid[r][c] = 0  # Dies or stays dead
                    
            else:  # Dead cell rules
                if neighbors == 3:
                    new_grid[r][c] = 1
            # print(f"Cell ({r}, {c}) has {neighbors} neighbors.")
    return new_grid


grid_indents = " "

def print_grid(grid):
    for row in grid:
        print(' '.join(['\033[92m█\033[0m' if cell else '·' for cell in row]))
    print()

def random_grid(rows, cols, alive_probability=0.3):
    rand_grid = []
    for r in range(rows):
        row = []
        rand_grid.append(row)
        for c in range(cols):
            if random.random() < alive_probability:
                row.append(1)
            else:
                row.append(0)
    return rand_grid
    # return [[1 if random.random() < alive_probability else 0 for _ in range(cols)] for _ in range(rows)]
```
